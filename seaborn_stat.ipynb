## parametric test

```python
from scipy.stats import shapiro,normaltest, anderson
def try_normal(data):
    result={'Anderson':{i:j for i,j in zip(anderson(data)[2],anderson(data)[1])},"Shapiro":shapiro(data)[1],'K^2':normaltest(data)[1]}
    result['Anderson']['stat']=anderson(data)[0]
    return result
```

```python
#Anderson is stat < critical value for accepting Ho when Ho - is normal distuted (or the designed dist)
anderson(a[a['trending_gap']<=5]['views'])
```

```python
try_normal(a[a['trending_gap']<=5]['views'])
```

```python
try_normal(np.log(a[a['trending_gap']<=5]['views']))
```

## Bootstrap and CDF

```python
def mean_differences(x1,x2,n_boot):
    mean_boot=[]
    for i in range (n_boot):
        mean_x1 = x1.sample(len(x1),replace = True).mean()
        mean_x2 = x2.sample(len(x2),replace = True).mean()
        dif_mean = mean_x1 - mean_x2
        mean_boot.append(dif_mean)
    hist, hist_edges = np.histogram(mean_boot, normed = True, bins = 10)
    a = np.cumsum(hist)
    width = hist_edges[1]-hist_edges[0]
    f1 = a*width
    plt.plot(hist_edges[1:],f1)
    plt.axvline(np.mean(x1)-np.mean(x2), color = 'r', linestyle = '--')
    plt.axvline(np.mean(x2)-np.mean(x1), color = 'b', linestyle = '--')
    plt.xticks(rotation = 90)
    plt.tight_layout()
    
```

```python
mean_differences(a[a['trending_gap']<=5]['views'],a[a['trending_gap']>5]['views'],1000)
```

## cek homogenitas (var sama atau tdk bs 2 klompok / lbh)

```python
from scipy.stats import levene, bartlett,f_oneway,kruskal
def try_equal_var(x1,x2):
    result={'Barlette:':{'Barlette-Stat': bartlett(x1,x2)[0], 'P-value':bartlett(x1,x2)[1]},
                        'Levene':{'Levene-stat':levene(x1,x2)[0],'P-value':levene(x1,x2)[1]}}
    return result
try_equal_var(youtube[youtube['category_id']==10]['views'],youtube[youtube['category_id']==29]['views'])
```

```python
bartlett(youtube[youtube['category_id']==10]['views'],youtube[youtube['category_id']==29]['views'],youtube[youtube['category_id']==20]['views'])
```

test barlette , p value = 0.00 < 0.05 maka Ho ditolak, jadi var beda.(jika data normal)

```python
levene(youtube[youtube['category_id']==10]['views'],youtube[youtube['category_id']==29]['views'],youtube[youtube['category_id']==20]['views'])
```

test levene , p value = 1.35e-11 < 0.05 maka Ho ditolak, jadi var beda.(jika data tidak normal= skew)

tidak normal dan var beda, sehingga di uji anova
